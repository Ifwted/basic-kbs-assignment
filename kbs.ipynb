This notebook demonstrates a simple Knowledge-Based System that stores facts, applies rules, and infers new knowledge using logical reasoning.


```python
class KnowledgeBase:
    def __init__(self):
        self.facts = set()
        self.rules = []

    def add_fact(self, fact):
        self.facts.add(fact)

    def add_rule(self, rule):
        self.rules.append(rule)

    def infer(self):
        new_facts = set()
        for rule in self.rules:
            if rule["condition"](self.facts):
                conclusion = rule["conclusion"]
                if conclusion not in self.facts:
                    new_facts.add(conclusion)
        self.facts.update(new_facts)
        return new_facts
```

```python
kb = KnowledgeBase()

# Adding a fact
kb.add_fact("It is cold")
```

```python
kb.add_rule({
    "condition": lambda facts: "It is cold" in facts,
    "conclusion": "People wear jackets"
})
```

```python
new_facts = kb.infer()

print("Newly inferred facts:", new_facts)
print("All facts in the knowledge base:", kb.facts)
```
